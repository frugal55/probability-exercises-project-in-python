- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

1. Tomar 3 blancas y 2 rojas.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

#print(ball_box)
            
# TODO

import random
import numpy as np
def cojo_bolas(s = 1000):
    count_1 = 0
    count_2 = 0
    for _ in range(1000):
        cb = []
        for _ in range(5):
            cb.append(ball_box[np.random.randint(0, 59)])
        cb = np.array(cb)

        bb = sum(cb == "white")
        br = sum(cb == "Red")
        bv = sum(cb == "Green")

        if (bb == 3) and ( br == 2):
            count_1 +=1
        if (bb == 5) or (br == 5) or (bv == 5):
            count_2 +=1
    return count_1 / 1000, count_2 / 1000

probabilidades = cojo_bolas(1000)
    

print(f"La probabilidad de tomar 3 bolas blancas y dos bolas rojas es de: {np.round(probabilidades[0] * 100, 2)}%")
print(f"la probabilidad de tomar las bolas del mismo color es: {np.round(probabilidades[1] * 100, 2)}%")





```
